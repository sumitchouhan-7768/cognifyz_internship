```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
data = pd.read_csv('Dataset  (1).csv')
```

```python
x1 = data['Has Online delivery']
y = data['Average Cost for two']
x2 = data['Has Table booking']
```

```python
plt.figure(figsize=(9,4))
plt.subplot(1,2,1)
sns.boxplot(x=x1,y=y)
plt.xlabel('Online delivery')
plt.ylabel('Average Cost for two')
plt.title('Online delivery affected by cost')
plt.subplot(1,2,2)
sns.boxplot(x=x2,y=y)
plt.xlabel('Tabel booking')
plt.ylabel('Average Cost for two')
plt.title('Table booking affected by cost')
plt.tight_layout()
plt.show()           
           
           
```

```python
#Restaurants having average cost les than 10,000
```

```python
plt.figure(figsize=(9,4))
plt.subplot(1,2,1)
sns.boxplot(x=x1, y='Average Cost for two', data = pd.read_csv('Dataset  (1).csv')[data['Average Cost for two']<10000])
plt.title('Online delivery by cost')
plt.subplot(1,2,2)
sns.boxplot(x=x2, y='Average Cost for two', data = pd.read_csv('Dataset  (1).csv')[data['Average Cost for two']<10000])
plt.title('Table booking by cost')
plt.tight_layout()
plt.show()
```
